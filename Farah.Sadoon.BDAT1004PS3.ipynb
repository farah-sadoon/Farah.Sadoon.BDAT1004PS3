```python
# Step 1. Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Step 2. Create 3 different series, each of length 100, as specified. 
array1 = np.random.randint(1, 4, 100)
series1 = pd.Series(array1, name = "series1")

array2 = np.random.randint(1, 3, 100)
series2 = pd.Series(array2, name = "series2")

array3 = np.random.randint(10000, 30000, 100)
series3 = pd.Series(array3, name = "series3")
```

```python
# Step 3. Create a DataFrame by joining the Series by column
housing = pd.concat([series1, series2, series3],axis=1)
housing
```

```python
# Step 4. Change the name of the columns to bdrs, bathrs, price_sqr_meter
housing = housing.rename(columns={'series1': 'bdrs', 'series2': 'bathrs', 'series3': 'price_sqr_meter'})
housing
```

```python
# Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
bigcolumn = housing.stack()
bigcolumn
```

```python
# Step 6. Oops it seems it is going only until index 99. Is this true?
bigcolumn.index[-1]
```

```python
# Step 7. Reindex the DataFrame so it goes from 0 to 299
bigcolumn.reset_index()
```

Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009. Don't forget to label your axes! 

Colour each dot according to its year. 

```python
rev_docs = pd.read_csv('/Users/farahsadoon/Desktop/arcade-revenue-vs-cs-doctorates.csv', header = 0)
colors = {2000:'C0', 2001:'C1', 2002:'C2', 2003:'C3', 2004:'C4', 2005:'C5', 2006:'C6', 2007:'C7', 2008:'C8', 2009:'C9'}
fig, ax = pp.subplots()

for i in range(len(rev_docs['Year'])):
    ax.scatter(rev_docs['Total Arcade Revenue (billions)'][i], 
               rev_docs['Computer Science Doctorates Awarded (US)'][i],
               color=colors[rev_docs['Year'][i]])

ax.set_title('Total Revenue Earned by Arcades by Number of Computer Science PhDs (U.S.)')
ax.set_xlabel('Total Arcade Revenue (billions)')
ax.set_ylabel('Number of Computer Science PhDs Awarded (US)')
```
